```python
import pandas as pd
```

# Pandemi öncesi ve sonrası İstanbuldaki evlerde kira değişimi

```python
df = pd.read_csv("kiraveri.csv")
pd.set_option('display.max_rows', None)
print(df)
```

# 2019 yılındaki 1+1 evlerin fiyatlarınının değişimi görüyoruz

```python
filtre = (df['YIL'] == 2019) & (df['ODA SAYISI'] == '1+1')
secilen_sutunlar = df.loc[filtre, ['YIL', 'ODA SAYISI',"AY",'ORT.FİYAT']]
secilen_sutunlar.head(5)
```

Bu analizde 2019 yılı içerisinde ortalama olarak 1+1 evlerin fiyatı aynı kalmış. Sadece pandeminin başlangıcı olan Mart ayında bir yükseliş görüyoruz.

# 2020 yılındaki 1+1 ev kira fiyatlarını inceleyelim

```python
filtre = (df['YIL'] == 2020) & (df['ODA SAYISI'] == '1+1')
secilen_sutunlar = df.loc[filtre, ['YIL', 'ODA SAYISI',"AY",'ORT.FİYAT']]
secilen_sutunlar.head(5)
```

Bu analiz sonucunda ise 2020 yılında da ev fiyatlarında bir önceki yıla göre artış görüyoruz. Her ay bir önceki aya göre fiyatlar artmaya devam etmiş. Şuan pandemi süreci devam ediyor.

2021 yılında ise önceki yıla göre artış devam etmiş. Özellikle yazın fiyatlar daha da artmış
